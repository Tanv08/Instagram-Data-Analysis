```python
import pandas as pd
import numpy as np
```

```python
df_insta = pd.read_csv(r'C:\Users\15512\Desktop\Marketing Analytics\instagram_global_top_1000.csv')
df_insta.head()
```

```python
# looking at info for dataframe

df_insta.info
```

```python
# using describe to get the basic statistics of the dataframe

df_insta.describe()
```

```python
# returning the relative frequency of audience country elements
# can see that 28% of the top 1,000 influencers have the United States as their audience country

df_insta['Audience Country'].value_counts(normalize=True).round(2)
```

```python
# creating a pivot table to show statistics for categories

df_insta.pivot_table(index="Category", values="Followers", aggfunc=['mean', 'sum', 'min', 'max'], sort=False)
```

```python
# creating a new dataframe just for the United States 

df_usa = df_insta[df_insta['Audience Country'] == 'United States']
df_usa.head()
```

```python
# turning the floats to ints

df_usa = df_usa.astype({"Followers":"int","Authentic engagement":"int","Engagement avg":"int"})
df_usa.head()
```

```python
# pivot table to show statistics on category followers in the United States

df_usa.pivot_table(index="Category", values="Followers", aggfunc=['mean', 'sum', 'min', 'max'], sort=False)
```

```python
df_usa.pivot_table(index="Category", values="Engagement avg", aggfunc=['mean', 'sum', 'min', 'max'], sort=False)
```

```python
# showing the categories with the most followers

topfollowers = df_usa.nlargest(10,'Followers')

topfollowers.pivot_table(index="Category", values="Followers", aggfunc=['sum'], sort=False)
```

```python
# creating dataframe for top 20 followers

top20followers = df_usa.nlargest(20,'Followers')
```

```python
# showing the categories with the most avg audience engagement

topengagementavg = df_usa.nlargest(20,'Engagement avg')

topengagementavg.pivot_table(index="Category", values="Engagement avg", sort=False)
```

```python
# showing the categories with the most authentic audience engagement

topengagement = df_usa.nlargest(20,'Authentic engagement')

topengagement.pivot_table(index="Category", values="Authentic engagement", sort=False)
```

```python
# creating a new fashion category dataframe to include all of the subsets

fashion = df_usa[df_usa['Category'].str.contains('Fashion', na=False)]
fashion.head()
```

```python
# import pyplot package

from matplotlib import pyplot as plt
```

```python
# plotting followers

plt.figure(figsize=(12,8))
fig1 = plt.bar(topfollowers.Category, topfollowers.Followers, label='Followers', width=0.8, color='skyblue')
plt.xticks(fontsize = 12, rotation = 90)
plt.ylabel("Followers", fontsize = 16)
plt.xlabel("Category", fontsize = 16)
plt.title("Top 10 Categories by Followers", fontsize = 20)
plt.tight_layout()
# remove scientific notation
plt.ticklabel_format(axis="y", style='plain')
```

```python
# plotting followers

plt.figure(figsize=(12,8))
fig2 = plt.bar(top20followers.Category, top20followers.Followers, label='Followers', width=0.8, color='skyblue')
plt.xticks(fontsize = 12, rotation = 90)
plt.ylabel("Followers", fontsize = 16)
plt.xlabel("Category", fontsize = 16)
plt.title("Top 20 Categories by Followers", fontsize = 20)
plt.tight_layout()
# remove scientific notation
plt.ticklabel_format(axis="y", style='plain')
```

```python
# plotting the number of accounts between 0 and 60 million followers

plt.figure(figsize=(12,8))
plt.hist(df_usa.Followers, color='skyblue', range=(0,60000000))
plt.xlabel('Followers', fontsize = 16)
plt.ylabel('Number of Instagram Accounts', fontsize = 16)
plt.title('Histogram: Follower Counts of Top Instagram Accounts in the US', fontsize = 20)
plt.tight_layout()
# remove scientific notation
plt.ticklabel_format(style='plain')
```

Analysis Insights

Found that 28% of the top 1,000 Instagram influencers have the United States as their audience country.

Interestingly, several of the top ranked Instagram accounts globally were under the category "Sports with a ball", but this category was not found in the top followers or audience engagement for the United States.

The top 10 follower categories in the United States were: Fashion|Modeling|Beauty, Music|Lifestyle, Music, Fashion|Beauty, Music|Fashion, Clothing|Outfits|Lifestyle, Modeling|Fashion, Nature landscapes|Science|Photography, Fitness|Gym|Shopping|Retail|Clothing|Outfits, Music|Cinema|Actors/actresses.

40% of these groups have "Fashion", 40% have "Music", 20% have "Modeling", 20% have "Beauty", 20% have "Lifestyle", 20% have "Clothing", and 20% have "Outfits".

The top 10 categories have a total of 2,458,500,000 followers. 42.66% of these are in a category involving "Fashion" and 41.57% are in a category involving "Music".

The top categories for engagement averages and authentic engagement were mostly consistent with the top followers. However, these categories also included "Cinema" and "Actors/Actresses".

Recommendations

Most of the top Instagram influencers are in a category of Fashion or Music. Many of the top categories also include Cinema, Acting, Modeling, Beauty, and Lifestyle. Categories involving Fashion had the highest followers and engagement. Therefore, Fashion should be considered the top category for Instagram influencers in the United States.
